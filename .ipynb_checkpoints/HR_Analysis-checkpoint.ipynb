```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

# importing data

```python
emp=pd.read_csv("HR_Analytics-Raw_Data.csv")
emp
```

```python
emp.info()
```

# data cleaning  

```python
pd.isnull(emp).sum()
```

```python
emp.dropna(inplace=True)
emp.head()
```

```python
pd.isnull(emp).sum()
```

```python
emp[emp.duplicated()]
```

```python
emp.drop_duplicates(inplace=True)
emp.head()
```

```python
emp[emp.duplicated()]
```

```python
emp.drop("DailyRate",axis=1,inplace=True)
emp.drop("Education",axis=1,inplace=True)
emp.drop("EmployeeCount",axis=1,inplace=True)
emp.drop("EmployeeNumber",axis=1,inplace=True)
emp.drop("HourlyRate",axis=1,inplace=True)
emp.drop("RelationshipSatisfaction",axis=1,inplace=True)
emp.drop("StockOptionLevel",axis=1,inplace=True)
emp.drop("YearsWithCurrManager",axis=1,inplace=True)
emp.drop("MonthlyIncome",axis=1,inplace=True)
emp.drop("MonthlyRate",axis=1,inplace=True)
emp.drop("NumCompaniesWorked",axis=1,inplace=True)
emp.drop("Over18",axis=1,inplace=True)
emp.drop("TrainingTimesLastYear",axis=1,inplace=True)
```

```python
emp.head()
```

```python
emp["Attrition"]=emp["Attrition"].replace("Yes",1)
emp["Attrition"]=emp["Attrition"].replace("No",0)
```

```python
emp.head()
```

```python
emp.info()
```

```python
emp["Attrition"]=emp["Attrition"].astype(int)
```
